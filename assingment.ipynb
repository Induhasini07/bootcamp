```python
# Student Performance Management System (SPMS)

students = []

# Function to calculate average marks
def calculate_average(marks):
    return sum(marks) / len(marks)

# Function to assign grade
def assign_grade(avg):
    if avg >= 75:
        return "Distinction"
    elif avg >= 60:
        return "First Class"
    elif avg >= 40:
        return "Second Class"
    else:
        return "Fail"

# Add new student
def add_student():
    roll = input("Enter Roll Number: ")

    for student in students:
        if student["roll"] == roll:
            print("Student with this Roll Number already exists!")
            return

    name = input("Enter Name: ")
    branch = input("Enter Branch: ")

    marks = []
    for i in range(3):
        mark = float(input(f"Enter Marks for Subject {i+1}: "))
        marks.append(mark)

    avg = calculate_average(marks)
    grade = assign_grade(avg)

    student = {
        "roll": roll,
        "name": name,
        "branch": branch,
        "marks": marks,
        "average": avg,
        "grade": grade
    }

    students.append(student)
    print("Student record added successfully!")

# View all students
def view_students():
    if not students:
        print("No records found.")
        return

    print("\n--- Student Records ---")
    for student in students:
        print(f"Roll No: {student['roll']}")
        print(f"Name: {student['name']}")
        print(f"Branch: {student['branch']}")
        print(f"Marks: {student['marks']}")
        print(f"Average: {student['average']:.2f}")
        print(f"Grade: {student['grade']}")
        print("-" * 30)

# Search student
def search_student():
    key = input("Enter Roll Number or Name: ")

    found = False
    for student in students:
        if student["roll"] == key or student["name"].lower() == key.lower():
            print("\nStudent Found:")
            print(student)
            found = True

    if not found:
        print("Student not found.")

# Update student record
def update_student():
    roll = input("Enter Roll Number to update: ")

    for student in students:
        if student["roll"] == roll:
            student["name"] = input("Enter New Name: ")
            student["branch"] = input("Enter New Branch: ")

            marks = []
            for i in range(3):
                mark = float(input(f"Enter New Marks for Subject {i+1}: "))
                marks.append(mark)

            student["marks"] = marks
            student["average"] = calculate_average(marks)
            student["grade"] = assign_grade(student["average"])

            print("Record updated successfully!")
            return

    print("Student not found.")

# Delete student
def delete_student():
    roll = input("Enter Roll Number to delete: ")

    for student in students:
        if student["roll"] == roll:
            students.remove(student)
            print("Record deleted successfully!")
            return

    print("Student not found.")

# Calculate class average
def class_average():
    if not students:
        print("No student records available.")
        return

    total = sum(student["average"] for student in students)
    avg = total / len(students)

    print(f"Class Average = {avg:.2f}")

# Assign grades to all students
def assign_grades_all():
    for student in students:
        student["grade"] = assign_grade(student["average"])

    print("Grades assigned successfully.")

# Filter students by grade
def filter_by_grade():
    grade = input(
        "Enter Grade (Distinction/First Class/Second Class/Fail): "
    )

    found = False
    for student in students:
        if student["grade"].lower() == grade.lower():
            print(student)
            found = True

    if not found:
        print("No students found in this category.")

# Find class topper
def find_topper():
    if not students:
        print("No records available.")
        return

    highest = max(student["average"] for student in students)

    print("\nClass Topper(s):")
    for student in students:
        if student["average"] == highest:
            print(
                f"{student['name']} "
                f"(Roll No: {student['roll']}) "
                f"- Average: {student['average']:.2f}"
            )

# Main menu
while True:
    print("\n===== Student Performance Management System =====")
    print("1. Add New Student")
    print("2. View All Students")
    print("3. Search Student")
    print("4. Update Student Record")
    print("5. Delete Student")
    print("6. Calculate Class Average")
    print("7. Assign Grades to All Students")
    print("8. Filter Students by Grade")
    print("9. Find Class Topper")
    print("10. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_student()
    elif choice == "2":
        view_students()
    elif choice == "3":
        search_student()
    elif choice == "4":
        update_student()
    elif choice == "5":
        delete_student()
    elif choice == "6":
        class_average()
    elif choice == "7":
        assign_grades_all()
    elif choice == "8":
        filter_by_grade()
    elif choice == "9":
        find_topper()
    elif choice == "10":
        print("Exiting SPMS...")
        break
    else:
        print("Invalid choice! Please try again.")
```
